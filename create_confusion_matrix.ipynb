```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
# Replace 'your_file.csv' with the path to your CSV file
model_pred = pd.read_csv('detection_results_main.csv')

model_pred.head()
```

```python
confusion_model = pd.crosstab(model_pred.Predicted_Class,model_pred.Actual_Class)
confusion_model = confusion_model.astype(int)
```

```python
fig = plt.figure(figsize=(17,5))
ax1 = plt.subplot(121)
sn.heatmap(confusion_model, annot=True, fmt='d', cmap="Blues")
```
